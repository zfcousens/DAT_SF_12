####7. Show 10 of the sidewalk cleanings in the Haight Ashbury neighborhood

```python
pd.read_sql("SELECT * FROM sf311 WHERE neighborhood = 'Haight Ashbury' AND category like '%Sidewalk Cleaning%' LIMIT 10", con)
```

####8. Which type of complaint was most common? 

```python
pd.read_sql("""SELECT category, COUNT(*) AS count FROM sf311 
                GROUP BY category ORDER BY count desc LIMIT 1""", con)
```

####9. Are there any neighborhoods where sidewalk cleaning requests aren't the most common? 

```python
pd.read_sql("""SELECT neighborhood, category FROM
                    (SELECT neighborhood, category, MAX(cat_count)   FROM 
                        (SELECT neighborhood, category, COUNT(category) AS cat_count 
                        FROM sf311 GROUP BY neighborhood, category)
                    GROUP BY neighborhood)
                WHERE category != 'Street and Sidewalk Cleaning'""", con)
```
