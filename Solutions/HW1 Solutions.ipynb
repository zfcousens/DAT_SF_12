####5. Turn the 311 data into a database. You can use the pandas read_sql command (example [here)]('http://bit.ly/1EoP81v') or the db package from yHat to do this. 

```python
# Write to Database.  We will then use to run queries
con = sqlite3.connect(":memory:")
df.to_sql('sf_311', con)
```

###### 6. Display the first 5 rows of the DB

```python
pd.read_sql('SELECT * FROM sf_311 LIMIT 5', con)
```

####7. Show 10 of the sidewalk cleanings in the Haight Ashbury neighborhood

```python
pd.read_sql('''SELECT * FROM sf_311 WHERE request_type = "Sidewalk_Cleaning" AND neighborhood = "Haight Ashbury" LIMIT 10 ''', con)
```

####8. Which type of complaint was most common? 

```python
pd.read_sql('SELECT request_type, count(*) FROM sf_311 GROUP BY request_type ORDER BY count(*) DESC LIMIT 10',con)
```

####9. Are there any neighborhoods where sidewalk cleaning requests aren't the most common? 

```python
pd.read_sql('SELECT request_type, neighborhood, count(*) FROM sf_311 GROUP BY request_type,neighborhood ORDER BY count(*) DESC LIMIT 10', con)
```
